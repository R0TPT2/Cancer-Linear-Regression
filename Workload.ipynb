```python
import numpy as np
import pandas as pd
```

```python
data_frame = pd.read_csv("cancer-habit-Linear_regression.csv")
```

```python
data_frame.info()
```

```python
data_frame.dropna(inplace=True)
```

```python
data_frame
```

```python
data_frame.info()
```

```python
data_frame = data_frame.drop("Sr No.", axis = 1)
```

```python
data_frame
```

```python
data_frame_c = pd.get_dummies(data_frame)
```

```python
data_frame_c
```

```python
from sklearn.model_selection import train_test_split

x = data_frame_c.drop(['Probability of Cancer'], axis = 1)
y = data_frame_c['Probability of Cancer']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.5)
```

```python
from sklearn.linear_model import LinearRegression


Li_Regression = LinearRegression()
Li_Regression.fit(x_train, y_train)
```

```python
Li_Regression.score(x_test, y_test)
```
